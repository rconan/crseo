```python
import numpy as np
import matplotlib.pyplot as plt;
plt.style.use("dark_background")
```

# On-axis wavefront in the exit pupil

```python
opd = np.load("on-axis.pkl",allow_pickle=True)
fig,ax = plt.subplots(figsize=(9,7))
h = ax.tripcolor(opd['x'],opd['y'],opd['opd'],shading='flat')
ax.set_aspect('equal')
ax.grid(alpha=0.25)
fig.colorbar(h,ax=ax)
```
